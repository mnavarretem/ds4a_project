```python
#File names and base path.
base_datapath = 'C:\\Users\\sapmn3\\Database\\OFFCORSS\\data\\OFFCORSS_Ventas'
file_list = ['Franquicias_2018',
             'Franquicias_2019',
             'tiendas_Diciembre_2019',
             'tiendas_Enero_2019',
             'tiendas_Noviembre_2019',
             'ventas_TV_2019'
            ]
file_list = ['Franquicias_2018',
             'Franquicias_2019',
             'tiendas_Diciembre_2019',
             'tiendas_Enero_2019',
             'tiendas_Noviembre_2019',
             'ventas_TV_2019'
            ]

data_ventas = []

for cur_file in file_list:
    file_source = cur_file.split('_')[0]
    print(file_source)
    cur_path = os.path.join(base_datapath, cur_file + '.csv')
    cur_data = pd.read_csv(cur_path, encoding = 'latin1', delimiter = ';')
        
    cur_data['Source'] = file_source
    
    cur_data.head(2)
    
    if len(data_ventas) == 0:
        data_ventas = cur_data
    else:
        data_ventas = pd.concat([data_ventas, cur_data], axis=0)

data_ventas
```
