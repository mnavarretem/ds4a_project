```python
import xml.etree.ElementTree as et 
import os
import pandas as pd
import numpy  as np
```

```python
# Read xml files ;


def read_tweets_xml(cur_path):
    xtree = et.parse(cur_path)
    xroot = xtree.getroot()

    df_cols = ["tweetid", "content", "lang", "sentiment"]
    rows = []

    for node in xroot: 
        s_tweetid   = node.find("tweetid").text if node is not None else None
        s_content   = node.find("content").text if node is not None else None
        s_lang      = node.find("lang").text if node is not None else None
        
        try:
            s_sentiment = node.find("sentiment").find("polarity").find("value").text if node is not None else None

        except:
            s_sentiment = node.find("sentiments").find("polarity").find("value").text if node is not None else None
            
            
        rows.append({"tweetid": s_tweetid, "content": s_content, 
                     "lang": s_lang, "sentiment": s_sentiment})

    out_df = pd.DataFrame(rows, columns = df_cols)
                          
    return out_df
```

```python
#File names and base path.
base_datapath = 'C:\\Users\\sapmn3\\Database\\OFFCORSS\\data\\MODEL'
file_list = ['TASS2019_country_CR_dev',
             'TASS2019_country_CR_train',
             'TASS2019_country_ES_train',
             'TASS2019_country_MX_dev',
             'TASS2019_country_MX_train',
             'TASS2019_country_PE_dev',
             'TASS2019_country_PE_train',
             'TASS2019_country_UY_dev',
             'TASS2019_country_UY_train',
             'TASS2019_country_ES_dev',
             'TASS2019_country_ES_dev',
             'intertass-CR-development-tagged',
             'intertass-CR-train-tagged',
             'intertass-ES-development-tagged',
             'intertass-ES-train-tagged',
             'intertass-PE-development-tagged',
             'intertass-PE-train-tagged',
             'general-train-tagged-3l',
             'general-test1k-tagged-3l',
             'general-test-tagged-3l',
            ]

data_tweets = []
for cur_file in file_list:
    print('Processing file ' + cur_file)
    cur_path = os.path.join(base_datapath, cur_file + '.xml')
    cur_data = read_tweets_xml(cur_path)
    if len(data_tweets) == 0:
        data_tweets = cur_data
    else:
        data_tweets = pd.concat([data_tweets, cur_data], axis=0)
        
data_tweets = data_tweets.reset_index()
```
