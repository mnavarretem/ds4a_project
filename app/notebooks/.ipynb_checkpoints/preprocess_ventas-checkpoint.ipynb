```python
# Load packages
import pandas as pd
import numpy  as np
from collections import Counter
import os
```

```python
#File names and base path.
base_datapath = 'C:\\Users\\sapmn3\\Database\\OFFCORSS\\data\\OFFCORSS_Ventas'

file_list = ['Franquicias_2019',
             'Franquicias_2020',
            ]

ventas_franquicias = []

for cur_file in file_list:
    file_source = cur_file.split('_')[0]
    print(cur_file)
    cur_path = os.path.join(base_datapath, cur_file + '.csv')
    cur_data = pd.read_csv(cur_path, encoding = 'latin1', delimiter = ';')
        
    cur_data['Source'] = file_source
        
    if len(ventas_franquicias) == 0:
        ventas_franquicias = cur_data
    else:
        ventas_franquicias = pd.concat([ventas_franquicias, cur_data], axis=0)
```

```python
#File names and base path.
base_datapath = 'C:\\Users\\sapmn3\\Database\\OFFCORSS\\data\\OFFCORSS_Ventas\\TiendasPropias'

file_list = ['Enero_2019','Febrero_2019','Marzo_2019','Abril_2019',
             'Mayo_2019','Junio_2019','Julio_2019','Agosto_2019',
             'Septiembre_2019','Octubre_2019','Noviembre_2019','Diciembre_2019',
             'Enero_2020','Febrero_2020','Marzo_2020','Abril_2020',
             'Mayo_2020','Junio_2020','Julio_2020','Agosto_2020',
            ]

ventas_tiendas = []

for cur_file in file_list:
    file_source = cur_file.split('_')[0]
    print(cur_file)
    cur_path = os.path.join(base_datapath, cur_file + '.csv')
    cur_data = pd.read_csv(cur_path, encoding = 'latin1', delimiter = ';')
        
    cur_data['Source'] = file_source
        
    if len(ventas_tiendas) == 0:
        ventas_tiendas = cur_data
    else:
        ventas_tiendas = pd.concat([ventas_tiendas, cur_data], axis=0)
```

```python
base_datapath = 'C:\\Users\\sapmn3\\Database\\OFFCORSS\\data\\OFFCORSS_Ventas'

file_list = ['TV_2019',
             'TV_2020'
            ]

ventas_virtual = []

for cur_file in file_list:
    file_source = cur_file.split('_')[0]
    print(cur_file)
    cur_path = os.path.join(base_datapath, cur_file + '.csv')
    cur_data = pd.read_csv(cur_path, encoding = 'latin1', delimiter = ';')
        
    cur_data['Source'] = 'tVirtual'
        
    if len(ventas_virtual) == 0:
        ventas_virtual = cur_data
    else:
        ventas_virtual = pd.concat([ventas_virtual, cur_data], axis=0)
        
```
