```python
# Load packages
import nltk # imports the natural language toolkit
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import string
import re
from collections import Counter
from nltk.stem import PorterStemmer
from nltk.util import ngrams
from pylab import rcParams
from wordcloud import WordCloud
import nltk
import os

from translate import Translator
```

```python
data_responses = pd.read_csv('SentiWordNet.csv', sep='\t')
data_responses = data_responses[(data_responses['PosScore'] > 0) | (data_responses['NegScore'] > 0) ]
data_responses
```

```python
data_responses['SynsetTerms'] = data_responses['SynsetTerms'].str.replace('_', ' ', regex=False)
data_responses['SynsetTerms'] = data_responses['SynsetTerms'].str.replace('#', ',', regex=False)
data_responses['SynsetTerms'] = data_responses['SynsetTerms'].str.replace('\d+', '', regex=True)
data_selected = data_responses['SynsetTerms'].reset_index(drop = True)
data_selected
len(data_selected['SynsetTerms'])
```

```python
translated_ls = []
translator    = Translator(to_lang="Spanish")
text_ls       = data_selected


for idx in range(8102,len(text_ls)):
    translation = translator.translate(text_ls.iloc[idx])
    translated_ls.append(translation)
    print('translating: {0:2.2f}% --> '.format(100 * idx/len(text_ls))+translation)
    
#translation = translator.translate('light up, clear up, clear, brighten,')
#print(translation)
```
